```python
import pandas as pd
import numpy as np

def prepare_data(M, Ci):

    rdata0 = pd.read_csv("BankChurners.csv")
    rdata = rdata0.drop([rdata0.columns[0],rdata0.columns[-1],rdata0.columns[-2]],axis=1)
    rdata.head()
    
    print (rdata.shape)
    rdata = rdata.drop([rdata.index[532],rdata.index[1225],rdata.index[2538],rdata.index[3332],rdata.index[5132]])
    print (rdata.shape)
    
    Dicts = []
    for c in range(len(rdata.columns)):
        l1 = rdata[rdata.columns[c]]
        l2 = list(dict.fromkeys(l1))
        l2.sort()
        nn = len(l2)
        Dicts.append(dict())
        for i in range(len(l2)):
            Dicts[c] [l2[i]] = i / nn 
    
    rM = 1./M
    
    XArray = np.zeros(  (rdata.shape[0], rdata.shape[1]-1),dtype = "int")
    YArray = np.zeros(  (rdata.shape[0], 1),dtype = "int")
 

            
    for n in range(rdata.shape[0]):
        for c in range(rdata.shape[1]):
            #print (rdata [n,c])
            if c < Ci:
                XArray[n,c] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM
            if c > Ci:
                XArray[n,c-1] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM
            if c == Ci:
                YArray[n,0] =  ( Dicts[c] [ rdata.iat [n, c]]) // rM
       
        
    return (XArray,YArray)

def prepare_tdata(M, Ci):

    rdata0 = pd.read_csv("BankChurners.csv")
    rdata = rdata0.drop([rdata0.columns[0],rdata0.columns[-1],rdata0.columns[-2]],axis=1)
    rdata.head()
    
    print (rdata.shape)
    #rdata = rdata.drop([rdata.index[532],rdata.index[1225],rdata.index[2538],rdata.index[3332],rdata.index[5132]])
    rdata = rdata.iloc[[532,1225,2538,3332,5132]]
    print (rdata.shape)
    
    Dicts = []
    for c in range(len(rdata.columns)):
        l1 = rdata[rdata.columns[c]]
        l2 = list(dict.fromkeys(l1))
        l2.sort()
        nn = len(l2)
        Dicts.append(dict())
        for i in range(len(l2)):
            Dicts[c] [l2[i]] = i / nn 
    
    rM = 1./M
    
    XArray = np.zeros(  (rdata.shape[0], rdata.shape[1]-1),dtype = "int")
    YArray = np.zeros(  (rdata.shape[0], 1),dtype = "int")
 

            
    for n in range(rdata.shape[0]):
        for c in range(rdata.shape[1]):
            #print (rdata [n,c])
            if c < Ci:
                XArray[n,c] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM
            if c > Ci:
                XArray[n,c-1] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM
            if c == Ci:
                YArray[n,0] =  ( Dicts[c] [ rdata.iat [n, c]]) // rM
       
        
    return (XArray,YArray)
```

```python
import numpy as np
Ci = 3
M = 10
(X, Y) =  prepare_data(M, Ci)
print (X.shape, Y.shape)
from sklearn.naive_bayes import CategoricalNB
clf = CategoricalNB()
clf.fit(X, Y)
#print(clf.predict(X[2:3]), Y[2:3])
YP = clf.predict(X)
dbok = 0
db  = 0
for n in range(X.shape[0]):
    if YP[n] == Y[n,0]:
        dbok = dbok + 1
    db = db + 1
print ("acc=", dbok/db )

# [532,1225,2538,3332,5132]

(X, Y) =  prepare_tdata(M, Ci)
print (X.shape, Y.shape)

YP = clf.predict(X)
dbok = 0
db  = 0
for n in range(X.shape[0]):
    if YP[n] == Y[n,0]:
        dbok = dbok + 1
    db = db + 1
print ("acc2=", dbok/db )
```
