```python
import pandas as pd
import numpy as np

from numpy import hstack
from numpy import zeros
from numpy import ones
from numpy.random import rand
from numpy.random import randn
from keras.models import Sequential
from keras.layers import Dense
from matplotlib import pyplot
from plot_keras_history import plot_history
import random 
import math

def define_discriminator(n_inputs):
	model = Sequential()
	model.add(Dense(8 *n_inputs, activation='relu', kernel_initializer='he_uniform', input_dim=n_inputs))
	model.add(Dense(8 *n_inputs, activation='relu', kernel_initializer='he_uniform', input_dim=n_inputs))
	model.add(Dense(1, activation='sigmoid'))
	# compile model
	model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
	return model

# define the standalone generator model
def define_generator(latent_dim, n_outputs):
	model = Sequential()
	model.add(Dense(6*(latent_dim +  n_outputs), activation='relu', kernel_initializer='he_uniform', input_dim=latent_dim))
	model.add(Dense(6*(latent_dim +  n_outputs), activation='relu', kernel_initializer='he_uniform', input_dim=latent_dim))
	model.add(Dense(n_outputs, activation='linear'))
	return model

# define the standalone generator model
def define_regenerator(latent_dim, n_outputs):
	model = Sequential()
	model.add(Dense(6*(latent_dim +  n_outputs), activation='relu', kernel_initializer='he_uniform', input_dim=latent_dim))
	model.add(Dense(6*(latent_dim +  n_outputs), activation='relu', kernel_initializer='he_uniform', input_dim=latent_dim))
	model.add(Dense(n_outputs, activation='linear'))
	return model

# define the combined generator and discriminator model, for updating the generator
def define_gan(generator, discriminator):
	# make weights in the discriminator not trainable
	discriminator.trainable = False
	# connect them
	model = Sequential()
	# add generator
	model.add(generator)
	# add the discriminator
	model.add(discriminator)
	# compile model
	model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
	return model

def define_regan(generator, regenerator):
	# make weights in the discriminator not trainable
	generator.trainable = False
	# connect them
	model = Sequential()
	# add generator
	model.add(regenerator)
	# add the discriminator
	model.add(generator)
	# compile model
	model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mse'])
	return model

def train_discrimantors(discriminators,DA, DD, N, M, N2, E, Ci):

    N1 = int(N2/3)

    Indx = [[] for _ in range(M)]
    TrainFalse  = [[] for _ in range(M)]
    
    for n in range(N):
        v = int(DA[n,Ci])
        Indx[v].append(n)
    
    for v in range(M):
        print(v,len(Indx[v]))
        
    
    for n in range(N):
        v = DA[n,Ci]
        for w in range(M):
            xu = []
            for i in range(data_dim):
                xu.append(DA[n,i])
            xu[Ci] = w
            if tuple (xu) not in DD.keys():
                xr = xu.copy()
                del xr[Ci]
                TrainFalse[w].append(tuple(xr))
        
    
    for v in range(M):
        print (v,":", len(TrainFalse[v]))
        
    
    for v in range(M):        
        
        if len(Indx[v]) == 0:
            continue
            
        x_real = np.zeros((N1,data_dim-1))
        y_real = np.zeros((N1,1))

        for n in range(N1):
            i = random.randint(0,len(Indx[v])-1)
            for j in range(data_dim):
                if j < Ci:
                    x_real[n,j] = DArray[Indx[v][i]][j] 
                if j > Ci:
                    x_real[n,j-1] = DArray[Indx[v][i]][j] 
            y_real[n,0] = 1


        x_fake = np.zeros((2*N1,data_dim-1))
        y_fake = np.zeros((2*N1,1))

        N12 = min(len(TrainFalse[v]),int(1.5*N1))
        
        
        for n in range(N12):
            for j in range(data_dim-1):
                x_fake[n,j] = TrainFalse[v][n][j]
            y_fake[n,0] = 0
        
        
        for n in range(N12,2*N1):
            fnd = 1
            while fnd == 1:
                x = []
                for j in range(data_dim):
                    x.append(random.randint(0,M-1))
                if tuple(x) not in DD.keys():
                    fnd = 0
            for j in range(data_dim-1):
                if j < Ci:
                    x_fake[n,j] = x[j]             
                if j > Ci:
                    x_fake[n,j-1] = x[j]             
            y_fake[n,0] = 0

        x_train = np.zeros((3*N1, data_dim-1))
        y_train = np.zeros((3*N1, 1))

        for n in range(N1):
            for j in range(data_dim-1):
                x_train[3*n,j] = x_real[n,j]
                x_train[3*n+1,j] = x_fake[n,j]
                x_train[3*n+2,j] = x_fake[N1+n,j]
            y_train [3*n,0] = 1
            y_train [3*n + 1,0] = 0
            y_train [3*n + 2,0] = 0


        r = discriminators[v].fit(x_train, y_train,epochs=E,verbose = 0)
        print ("v=",v)
        #plot_history(r.history)

def prepare_data(M, N0):

    rdata0 = pd.read_csv("BankChurners.csv")
    rdata = rdata0.drop([rdata0.columns[0],rdata0.columns[-1],rdata0.columns[-2]],axis=1)
    rdata.head()
    
    Dicts = []
    for c in range(len(rdata.columns)):
        l1 = rdata[rdata.columns[c]]
        l2 = list(dict.fromkeys(l1))
        l2.sort()
        nn = len(l2)
        Dicts.append(dict())
        for i in range(len(l2)):
            Dicts[c] [l2[i]] = i / nn 
    
    rM = 1./M

    N0 = min(N0,int(0.9*rdata.shape[0])) 
    N1 = rdata.shape[0] - N0
    
    DArray = np.zeros(  (rdata.shape[0], rdata.shape[1]),dtype = "float")
    DD = dict()
    TArray = np.zeros(  (N1, rdata.shape[1]),dtype = "float")

    #Tr = random.sample(range(rdata.shape[0]), N0)
    #Tt = []
    #for n in range(rdata.shape[0]):
    #    if n not in Tr:
    #        Tt.append(n)
            
    for n in range(rdata.shape[0]):
        for c in range(rdata.shape[1]):
            #print (rdata [n,c])
            DArray[n,c] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM
        DD[tuple(DArray[n,:]) ] = 1
       
    #for n in range(len(Tt)):
    #    for c in range(rdata.shape[1]):
    #        #print (rdata [n,c])
    #        TArray[n,c] =  ( Dicts[c] [ rdata.iat [Tt[n],c]]) // rM
    #    DD[tuple(TArray[n,:]) ] = 1
        
        
    print (DArray.shape,  len(DD), TArray.shape)
    print (DArray[1:3,:])
    return (DArray,DD, TArray)


def test_discriminator(discriminator, data_dim, DD, DA,  M, N1 ):
    
    db  = 0
    dbp = 0
    for _ in range(3):
        x_test = np.zeros((N1, data_dim))
        y_test = np.zeros((N1, 1))
    
        for n in range(N1):
            if random.random() < 0.1:
                m = random.randint(0, DA.shape[0]-1)
                for j in range(data_dim):
                    x_test[n,j] = DA[m,j]
                y_test[n,0] = 1
                dbp = dbp + 1
            else:
                for j in range(data_dim):
                    x_test[n,j] = random.randint(0,M-1)
                if tuple(x_test[n,:]) in DD.keys():
                    y_test[n,0] = 1
                    dbp = dbp + 1
                else:
                    y_test[n,0] = 0

        y_pred = discriminator.predict(x_test)

        for n in range(N1):
            if abs(y_pred[n]-y_test[n]) < .5:
                db = db + 1
           
    print ("accuracy=", db / (3*N1), dbp / (3*N1) )
    
def generate_latent_points(latent_dim, n):
	# generate points in the latent space
	x_input = randn(latent_dim * n)
	# reshape into a batch of inputs for the network
	x_input = x_input.reshape(n, latent_dim)
	return x_input    


def train_gans(DArray, gan_models, latent_dim, N3, E2, Ci):
    # prepare points in latent space as input for the generator
    
    x_train = generate_latent_points(latent_dim, N3)
    # create inverted labels for the fake samples
    y_train = ones((N3, 1))
    # update the generator via the discriminator's error
    for v in range(M):
        r = gan_models[v].fit(x_train, y_train,epochs=E2,verbose = 0)
        #plot_history(r.history) 
    
    
def train_regans(DArray, regan_models,  N4, E4, Ci):

    
    x_real = np.zeros((N4,data_dim-1))
    y_real = np.zeros((N4,data_dim-1))

    
    for v in range(M):
        
        N1 = min(DArray.shape [0], int(.5*N4))
        
        ndb = 0
        for n in range(DArray.shape[0]):
            if int(DArray[n,Ci]) == v:
                for j in range(data_dim):
                    if j < Ci:
                        x_real[ndb,j] = DArray[n][j] 
                        y_real[ndb,j] = DArray[n][j] 
                    if j > Ci:
                        x_real[ndb,j-1] = DArray[n][j] 
                        y_real[ndb,j-1] = DArray[n][j] 
                ndb = ndb + 1
                if ndb == N1:
                    break
        N1 = ndb
        for n in range(N1,N4):
            for j in range(data_dim-1):
                x_real[n,j] = random.randint(0,M-1)
                y_real[n,j] = x_real[n,j] 

           
        r = regan_models[v].fit(x_real, y_real,epochs=E4,verbose = 0)
        plot_history(r.history) 
    
            
M = 10    # values

(DArray, DDict, TArray)  = prepare_data(M, 9000)
N = DArray.shape[0]
data_dim = DArray.shape[1]
latent_dim = 14

N2 = 100000  # train size
E = 20  # Epochs
N3 = 300   # test size
Ci = 3  # attribute index

print ("Discriminator training")

discriminators = []
for v in range(M):
    discriminators.append(define_discriminator(data_dim-1))

print ("M:",M,"dim:",data_dim-1, "cube size:", M**(data_dim-1), "real size:",N, " train_size:", N2,"epochs:", E, "test_size:", N3, "mode: real data")
train_discrimantors(discriminators, DArray, DDict, N,M, N2, E, Ci)

N3 = 50000 # train size
E3 = 20 # epochs2 


generators = []
gan_models = []
for v in range(M):
    generators.append( define_generator(latent_dim,data_dim-1))
    gan_models.append(define_gan(generators[v], discriminators[v]))

print ("Gan , dim:", data_dim-1, "N size=",N3, "epochs:",E3)
    
train_gans(DArray, gan_models, latent_dim,N3, E3, Ci)


regenerators = []
for v in range(M):
    regenerators.append(define_regenerator(data_dim-1, latent_dim) )

regan_models = []
for v in range(M):
    regan_models.append( define_regan(generators[v], regenerators[v]))

    
N4 = 50000
E4 = 20
print ("train re-gan ", "N = ", N4, " epochs=", E4)
train_regans(DArray, regan_models, N4, E4, Ci)    

#test_discriminator(discriminator, data_dim, DDict, DArray, M, N3)
```
