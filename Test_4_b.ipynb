```python
import pandas as pd
import numpy as np
import random

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

def prepare_data(M, Nt, Ci):

    rdata0 = pd.read_csv("BankChurners.csv")
    rdata = rdata0.drop([rdata0.columns[0],rdata0.columns[-1],rdata0.columns[-2]],axis=1)
    rdata.head()
    #print (min( rdata[rdata.columns[Ci]]), max(rdata[rdata.columns[Ci]]))
    
    if Nt > 1000:
        Nt = 1000
        
    fbe = open("INDX.TXT","r")
    data = fbe.read().split("\n")
    inx2 = [int(x) for x in data if len(x) > 0]
    fbe.close()    
    indx = inx2[:Nt]

    #indx = random.sample(range(rdata.shape[0]), Nt)
    
    
    rindx = []
    for i in indx:
        rindx.append(rdata.index[i])
        
    print (rdata.shape)
    rdata = rdata.drop(rindx)
    print (rdata.shape)
    
    
    Dicts = []
    for c in range(len(rdata.columns)):
        l1 = rdata[rdata.columns[c]]
        l2 = list(dict.fromkeys(l1))
        l2.sort()
        nn = len(l2)
        Dicts.append(dict())
        for i in range(len(l2)):
            Dicts[c] [l2[i]] = i / nn 
    
    rM = 1./M

    DArray = np.zeros(  (rdata.shape[0], rdata.shape[1]),dtype = "float")
    
    for n in range(rdata.shape[0]):
        for c in range(rdata.shape[1]):
            #print (rdata [n,c])
            DArray[n,c] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM

        
    print ('A', DArray.shape)
           
    ###########################################
    
    
    rdata0 = pd.read_csv("BankChurners.csv")
    rdata = rdata0.drop([rdata0.columns[0],rdata0.columns[-1],rdata0.columns[-2]],axis=1)
    rdata.head()
    
    indx = random.sample(range(rdata.shape[0]), Nt)
    rindx = []
    for i in indx:
        rindx.append(i)
        
    print (rdata.shape)
    rdata = rdata.iloc[rindx]
    print (rdata.shape)
    
    
    Dicts = []
    for c in range(len(rdata.columns)):
        l1 = rdata[rdata.columns[c]]
        l2 = list(dict.fromkeys(l1))
        l2.sort()
        nn = len(l2)
        Dicts.append(dict())
        for i in range(len(l2)):
            Dicts[c] [l2[i]] = i / nn 
    
    print (Dicts[Ci])
    rM = 1./M

    TArray = np.zeros(  (rdata.shape[0], rdata.shape[1]),dtype = "float")
    Y = np.zeros(  (rdata.shape[0], 1),dtype = "float")
    
    for n in range(rdata.shape[0]):
        for c in range(rdata.shape[1]):
            #print (rdata [n,c])
            TArray[n,c] =  ( Dicts[c] [ rdata.iat [n,c]]) // rM
        Y[n] = TArray[n,Ci]
        TArray[n,Ci] = np.nan
        
        
    print ('B', DArray.shape,  TArray.shape)
    return (DArray,TArray,Y)


M = 10    # values
Ci = 3    # investigated attribute
N3 = 1000   # test size

(DArray,  TArray, TY)  = prepare_data(M, N3, Ci)
N = DArray.shape[0]
data_dim = DArray.shape[1]


imp_mean = IterativeImputer(random_state=0)
imp_mean.fit(DArray)

N1 = TArray.shape[0]
y_pred = imp_mean.transform(TArray)
db = 0
for n in range(N1):
    if abs(y_pred[n,Ci] - TY[n,0]) < 1:
        db = db + 1
print ("accuracy =", db / N1)
```
